```python
import numpy as np
import tensorflow as tf

XY = np.loadtxt("XOR.csv", delimiter = " ", dtype = np.float32)
x_data = XY[:,0:-1]
y_data = XY[:,[-1]]

print(type(x_data))
print(type(y_data))

X = tf.placeholder(tf.float32, shape = [None, 2])
Y = tf.placeholder(tf.float32, shape = [None, 1])

'''
Logistic Model
W = tf.Variable(tf.random_normal([2, 1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

hypothesis = tf.sigmoid(tf.matmul(X, W)+ b)
'''
with tf.name_scope("layer1") as scope:
    W1 = tf.Variable(tf.random_normal([2, 8]), name = 'weight1') # 여기서 output의 개수는 임의의 값
    b1 = tf.Variable(tf.random_normal([8]), name = 'bias1')
    layer1 = tf.sigmoid(tf.matmul(X, W1) + b1)

    w1_hist = tf.summary.histogram("weight1", W1)
    b1_hist = tf.summary.histogram("bias1", b1)
    layer1_hist = tf.summary.histogram("layer1", layer1)
    
with tf.name_scope("layer2") as scope:   
    W2 = tf.Variable(tf.random_normal([8, 5]), name = 'weight2')
    b2 = tf.Variable(tf.random_normal([5]), name = 'bias2')
    layer2 = tf.sigmoid(tf.matmul(layer1, W2) + b2)
    
    w2_hist = tf.summary.histogram("weight2", W2)
    b2_hisy = tf.summary.histogram("bias2", b2)
    layer2_hist = tf.summary.histogram("layer2", layer2)

with tf.name_scope("layer3") as scope:
    W3 = tf.Variable(tf.random_normal([5, 3]), name = 'weight3')
    b3 = tf.Variable(tf.random_normal([3]), name = 'bias3')
    layer3 = tf.sigmoid(tf.matmul(layer2, W3) + b3)

    w3_hist = tf.summary.histogram("weight3", W3)
    b3_hist = tf.summary.histogram("bias3", b3)
    layer3_hist = tf.summary.histogram("layer3", layer3)
    
with tf.name_scope("layer4") as scope:
    W4 = tf.Variable(tf.random_normal([3, 1]), name = 'weight4')
    b4 = tf.Variable(tf.random_normal([1]), name = 'bias4')
    hypothesis = tf.sigmoid(tf.matmul(layer3, W4) + b4)
    
    w4_hist = tf.summary.histogram("weight4", W4)
    b4_hist = tf.summary.histogram("bias4", b4)
    hypothesis_hist = tf.summary.histogram("hypothesis", hypothesis)

cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1-Y)*tf.log(1-hypothesis))
cost_summ = tf.summary.histogram("cost", cost)
train = tf.train.GradientDescentOptimizer(learning_rate = 0.05).minimize(cost)

predicted = tf.cast(hypothesis > 0.5, dtype = tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype = tf.float32))

'''
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
'''


with tf.Session() as sess:
    summary = tf.summary.merge_all()
    sess.run(tf.global_variables_initializer())
    #writer = tf.summary.FileWriter("./board/XOR_with_NN", sess.graph)
    writer = tf.summary.FileWriter("./board/XOR_with_NN")
    writer.add_graph(sess.graph)
    
    for step in range(2000+1):
        sess.run(train, feed_dict = {X : x_data, Y : y_data})
        #writer.add_summary(s, global_step = step)
        
        if step % 100 == 0 :
            #global_step = step
            print(step, sess.run(cost, feed_dict={X:x_data, Y:y_data}))
        '''
    s = sess.run(summary, feed_dict = {X :x_data, Y : y_data})
    writer.add_summary(s, global_step = global_step)
        '''
    h, c, a = sess.run([hypothesis, predicted, accuracy], feed_dict={X:x_data, Y:y_data})
    print("\nHypothesis: ", h, "\nCorrect: ", c, "\nAccuracy: ", a)
```
