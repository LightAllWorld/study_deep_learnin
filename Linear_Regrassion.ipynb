```python
import tensorflow as tf
import numpy as np

x_data = [1.0, 2.0, 3.0]
y_data = [1.0, 2.0, 3.0]

X = tf.placeholder(np.float32, shape = [None])
Y = tf.placeholder(np.float32, shape = [None])

W = tf.Variable(tf.random_normal([1]), name = "weight")
b = tf.Variable(tf.random_normal([1]), name = "bias")

hypothesis = X * W + b

cost = tf.reduce_mean(tf.square(hypothesis - Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.03)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

predicated = tf.cast( hypothesis > 0.5, dtype = tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicated, Y), dtype = tf.float32))

for step in range(2001):
    cost_val, _ = sess.run([cost, train], feed_dict = {X:x_data, Y:y_data})
    if step %20 == 0 :
        print(step, "\ncost :", cost_val, ", weight :",W_val, ", bias :", b_val)
        
h, c = sess.run([hypothesis, cost],
                  feed_dict ={X: [4.0, 5.0, 6.0], Y : [4.0, 5.0, 6.0]})
print("\nHypothesis: ", h, "\nCorrect (Y): ", c)
```
